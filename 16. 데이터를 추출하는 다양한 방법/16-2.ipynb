```python
import pandas as pd
df = pd.DataFrame({'var1' : [1,2,3],
                   'var2' : [4,5,6]})
df
```

```python
# 인덱스 문자열 지정하기
df = pd.DataFrame({'var1': [1,2,3] ,
                   'var2': [4,5,6]},
                  index = ['Kim', 'Lee', 'Park'])
df
```

```python
# Kim행 추출
df.loc['Kim']
```

```python
# Kim, Park행 추출
df.loc[['Kim','Park']]
```

```python
# Lee행의 var1열 추출
df.loc['Lee', 'var1']
```

```python
# Kim, Lee행의 var2열 추출
df.loc[['Kim','Lee'], 'var2']
```

```python
df.loc[, 'var1']
```

```python
df.loc[:, 'var1']
```

```python
# 인덱스 번호가 있는 데이터 프레임
import pandas as pd
df_raw = pd.read_csv('exam.csv')
df = df_raw.head(10)
df = df_raw.copy()
df
```

```python
# 0행 추출
df.loc[0]
```

```python
# 2행 추출
df.loc[2]
```

```python
# 데이터 프레임으로 추출
df.loc[[0]]
```

```python
# 1,3,5행 추출
df.loc[[0,3,5]]
```

```python
# 0~3행 추출
df.loc[0:3]
```

```python
# 7~9행 추출
df.loc[7:9]
```

```python
df.loc[:2]
```

```python
# 5행 ~ 마지막 행 추출
df.loc[5:]
```

```python
# 첫번째 행 ~ 마지막 행 추출
df.loc[:]
```

```python
# 인덱스 문자열이 있는 데이터 프레임
df_label = pd.DataFrame({'var1' : [1,2,3],
                         'var2' : [4,5,6]},
                        index = ['Kim', 'Lee', 'Park'])
df_label
```

```python
# 인덱스 번호로 행 추출하기
df_label.loc[0]
```

```python
# 인덱스 번호로 행 추출하기
df.loc[2, :]
```

```python
# 인덱스 번호로 열 추출하기
df.loc[:, 2]
```

```python
# nclass가 1이면 추출
df.loc[df['nclass'] == 1]
```

```python
# 수학 점수가 60점 이상이면 추출
df.loc[df['math'] >= 60]
```

```python
# nclass가 1인 행의 math, english열 추출
df.loc[df['nclass'] == 1, ['math','english']]
```

```python
# 0~3행의 math열 추출
df.loc[0:3, 'math']
```
